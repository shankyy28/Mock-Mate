```python
import numpy as np
import pandas as pd

df = pd.read_csv('D:\\Projects\\AI Interview Simulator\\MockMate\\Resume Parser\\UpdatedResumeDataSet.csv')
print(df)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

df['resume'] = df['resume'].str.lower()
df['resume'] = df['resume'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df['resume'] = df['resume'].apply(word_tokenize)
stop_words = set(stopwords.words('english'))
df['resume'] = df['resume'].apply(lambda x: [word for word in x if word not in stop_words])
df['resume'] = df['resume'].apply(lambda x: ' '.join(x))
#df = df.drop_duplicates()
#df = df.dropna()

df.to_csv('preprocessed_resume_dataset.csv', index=False)
print(df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
x=df['resume']
y=df['category']
vectorizer = TfidfVectorizer()
x_vectorized = vectorizer.fit_transform(x)
print(x_vectorized)


```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_vectorized, y, test_size=0.2, random_state=42)
print("x_train",x_train)
print("x_test",x_test)
print("y_train",y_train)
print("y_test",y_test)
```

```python
from sklearn.svm import SVC
svm_classifier = SVC()

# Train the classifier on your training data
svm_classifier.fit(x_train, y_train)
```

```python
import joblib
joblib.dump(svm_classifier, 'svm_model.joblib')
joblib.dump(vectorizer, 'tfidf_vectorizer.joblib')
```

```python
import spacy
import tkinter as tk
from tkinter import filedialog, messagebox, ttk
import PyPDF2
import joblib
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Load the spaCy NER model for extracting skills
nlp = spacy.load('en_core_web_sm')

# Load your model and vectorizer
loaded_classifier = joblib.load('svm_model.joblib')
loaded_vectorizer = joblib.load('tfidf_vectorizer.joblib')

# Define stop words
stop_words = set(stopwords.words('english'))

# List of technical skills (this could also be loaded from a file)
technical_skills = [
    'python', 'java', 'c++', 'html', 'css', 'javascript', 'react', 'angular', 'django',
    'flask', 'docker', 'kubernetes', 'aws', 'azure', 'git', 'sql', 'mysql', 'postgresql',
    'mongodb', 'tensorflow', 'pytorch', 'keras', 'linux', 'bash', 'shell', 'scala', 'ruby',
    'hadoop', 'spark', 'pandas', 'numpy', 'scikit-learn','testing','automation','flutter','communication','teamwork',
    'leadership'
]

def preprocess_resume(resume_text):
    resume_text = resume_text.lower()
    resume_text = resume_text.translate(str.maketrans('', '', string.punctuation))
    resume_text = word_tokenize(resume_text)
    resume_text = [word for word in resume_text if word not in stop_words]
    return resume_text

def extract_technical_skills(resume_tokens):
    # Extract technical skills that match the predefined list
    skills_in_resume = [skill for skill in resume_tokens if skill in technical_skills]
    return skills_in_resume[:5]  # Return the top 5 technical skills

def predict_job_profile(resume_text):
    # Preprocess the input resume text
    preprocessed_resume = preprocess_resume(resume_text)
    resume_vectorized = loaded_vectorizer.transform([' '.join(preprocessed_resume)])
    # Predict the job category using the loaded classifier
    predicted_category = loaded_classifier.predict(resume_vectorized)[0]
    return predicted_category

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            resume_text = ''
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                resume_text += page.extract_text()
            return resume_text
    except Exception as e:
        messagebox.showerror("Error", f"Failed to read PDF: {e}")
        return None

def upload_and_predict():
    pdf_path = filedialog.askopenfilename(filetypes=[("PDF Files", "*.pdf")])
    if pdf_path:
        resume_text = extract_text_from_pdf(pdf_path)
        if resume_text:
            # Preprocess the resume text
            resume_tokens = preprocess_resume(resume_text)

            # Predict the job profile
            predicted_category = predict_job_profile(' '.join(resume_tokens))

            # Extract top 5 technical skills
            top_technical_skills = extract_technical_skills(resume_tokens)

            # Display predicted job profile and top technical skills
            label_predicted_job.config(text=f"Predicted Job Profile: {predicted_category}")
            label_top_skills.config(text=f"Top 5 Skills: {', '.join(top_technical_skills)}")

# GUI: Setup Tkinter window
window = tk.Tk()
window.title("Resume Prediction")

frame = ttk.Frame(window, padding=20)
frame.pack(pady=20)

button_upload_pdf = ttk.Button(frame, text="Upload PDF Resume", command=upload_and_predict)
button_upload_pdf.grid(row=0, column=0, padx=5, pady=5)

label_predicted_job = ttk.Label(frame, text="")
label_predicted_job.grid(row=1, column=0, padx=5, pady=5)

# Add a label for displaying top 5 technical skills
label_top_skills = ttk.Label(frame, text="")
label_top_skills.grid(row=2, column=0, padx=5, pady=5)

# Run the tkinter event loop
window.mainloop()
```
